```python
df1=pd.get_dummies(df[['Fuel_Type','Selling_type','Transmission',]],drop_first=True)
df1
```

```python
dfe=pd.concat([df,df1],axis=1)
dfe
```

```python
dfe=dfe.drop(['Car_Name','Fuel_Type','Transmission','Selling_type'],axis=1)
dfe
```

```python
dfe.columns
```

Splitting data into independent and dependentvariables

```python
x=dfe.drop(['Present_Price'],axis=1)
```

```python
x
```

```python
y=dfe['Present_Price']
```

```python
y
```

Splitting date into training data and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

ModelCreation

Linear Regression

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual _value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_test,y_pred,color='blue')
```

```python
print("slope is",model.coef_)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

Performance Evaluation

```python
print("MAE",mean_absolute_error(y_test,y_pred))
print("Error percentage is ",mean_absolute_percentage_error(y_test,y_pred))
mean_sq_error=mean_squared_error(y_test,y_pred)
print("Mean squared error is ",mean_sq_error)
print("Root Mean Sqaured Error is ",np.sqrt(mean_sq_error))
print('r2 score is',r2_score(y_test,y_pred))
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_test,y_pred,color='blue')
```

Performance Evaluation

```python
print("MAE",mean_absolute_error(y_test,y_pred))
print("Error percentage is ",mean_absolute_percentage_error(y_test,y_pred))
mean_sq_error=mean_squared_error(y_test,y_pred)
print("Mean squared error is ",mean_sq_error)
print("Root Mean Sqaured Error is ",np.sqrt(mean_sq_error))
print('r2 score is',r2_score(y_test,y_pred))
```
