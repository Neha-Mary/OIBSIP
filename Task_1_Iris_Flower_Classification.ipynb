```python
df=df.drop(['Id'],axis=1)
df
```

Splitting data into independent and dependent variables

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

Splitting data into training and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=30)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

Normalization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

Model Creation

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

Performance Evaluation

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
print('Training Score',knn.score(x_train,y_train))
print('Testing Score',knn.score(x_test,y_test))
```

Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
bayes=GaussianNB ()
bayes.fit(x_train,y_train)
y_pred=bayes.predict(x_test)
y_pred
```

Performance Evaluation

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
print('Training Score',bayes.score(x_train,y_train))
print('Testing Score',bayes.score(x_test,y_test))
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dec=DecisionTreeClassifier(random_state=42)
dec.fit(x_train,y_train)
y_pred=dec.predict(x_test)
y_pred
```

Performance Evaluation

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
print('Training Score',dec.score(x_train,y_train))
print('Testing Score',dec.score(x_test,y_test))
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
y_pred
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
print('Training Score',rf.score(x_train,y_train))
print('Testing Score',rf.score(x_test,y_test))
```
