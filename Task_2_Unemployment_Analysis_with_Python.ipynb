```python
df[' Date'] = pd.to_datetime(df[' Date'])
df['Month']=df[' Date'].dt.month_name()
```

```python
df=df.drop([' Date'],axis=1)
```

```python
df.columns
```

```python
corrmat = df.corr()
plt.figure(figsize=(10,10))
g=sns.heatmap(corrmat,annot=True)
```

```python
df= df.drop([' Frequency','longitude','latitude'],axis=1)
```

State Wise Estimations

```python
sns.pairplot(df,hue='Region')
```

```python
sns.barplot(x=' Estimated Employed',y ='Region',data=df)
```

```python
sns.barplot(x =' Estimated Labour Participation Rate (%)',y= 'Region',data = df)
```

```python
sns.barplot(x= ' Estimated Unemployment Rate (%)',y = 'Region',data=df)
```

```python
plt.figure(figsize=(30,15))
plt.title('Estimated Employed')
sns.histplot(x= ' Estimated Employed', hue ='Region',data =df)
```

```python
plt.figure(figsize=(30,15))
plt.title(' Estimatted Labour Participation Rate')
sns.histplot(x= ' Estimated Labour Participation Rate (%)', hue ='Region',data =df)
```

```python
plt.figure(figsize=(30,15))
plt.title('Estimated Unemployment Rate')
sns.histplot(x= ' Estimated Unemployment Rate (%)', hue ='Region',data =df)
```

```python
sns.relplot(data=df, x=" Estimated Labour Participation Rate (%)", y=" Estimated Unemployment Rate (%)", hue="Region", kind="line")
```

```python
sns.relplot(data=df, x="Month", y=" Estimated Unemployment Rate (%)", hue="Region", kind="line")
```

```python
plt.figure(figsize =  (15,10))
data = df.groupby('Region')[' Estimated Labour Participation Rate (%)'].sum()
data.plot(kind='pie', autopct='%1.1f%%')
plt.title('Labour Rate by State')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize =  (15,10))
data = df.groupby('Region')[' Estimated Unemployment Rate (%)'].sum()
data.plot(kind='pie', autopct='%1.1f%%')
plt.title('Estimated Unemployment Rate by State')
plt.axis('equal')
plt.show()
```
